## Newton's Method

Recall that in logistic regression, the hypothesis function is 

$h_{\theta}(x) = g(\theta^{T} x) = \dfrac{1}{1+e^{-\theta^{T} x}} = P(y=1|x;\theta)$

````octave
g = inline('1.0 ./ (1.0 + exp(-z))'); 
% Usage: To find the value of the sigmoid 
% evaluated at 2, call g(2)
````

## Cost Function

The cost function $J(\theta)$ is defined as

$J(\theta) = - \dfrac{1}{m} \big[ \sum_{i=1}^{m} y^{(i)}log\big( h_{\theta}(x^{(i)}) \big) + \big(1-y^{(i)}\big) log\big(1-h_{\theta}(x^{(i)})\big)  \big]$

$\dfrac{\partial}{\partial \theta_j} J(\theta) = \dfrac{1}{m} \sum_{i=1}^{m} \big( h_{\theta}(x^{(i)}) - y^{(i)} \big) x^{(i)}_j$

$\theta_j \leftarrow \theta_j - \alpha \dfrac{\partial}{\partial \theta_j} J(\theta)$

__Matrix Version__

Our goal is to use Newton's method to minimize this function. Recall that the update rule for Newton's method is 

$\theta^{(t+1)} = \theta^{(t)}-H^{-1}\nabla_{\theta}J$

In logistic regression, the gradient and the Hessian are 

$\nabla_{\theta}J = \dfrac{1}{m} \sum_{i=1}^{m} \big( h_{\theta}(x^{(i)}) - y^{(i)} \big) x^{(i)}_j$

$H = \frac{1}{m}\sum_{i=1}^{m} \big[ h_{\theta}(x^{(i)}) \big( 1 - h_{\theta}(x^{(i)}) \big) x^{(i)} \big(x^{(i)}\big)^{T}\big]$

__Decision Boundary:__

The decision boundary is defined as the line where

$P(y=1 \vert x; \theta) = 0.5 \;\;\;\;\; \Longrightarrow \;\;\;\;\; \theta^T x = 0$

When

$\theta^T x = \begin{bmatrix}
    \theta_{0} & \theta_{1} & \theta_{2}
\end{bmatrix} \begin{bmatrix}
    1 \\
    x_{1}  \\
    x_{2}  \\
\end{bmatrix} =
\theta_0 + \theta_1 x_1 + \theta_2 x_2 = 0$

when $\theta_0 + \theta_1 x_1 + \theta_2 x_2 = 0$ is the middle(decision boundary line),

We can pick any get two points from the $x_1$:

```
x1 = c(min(x[,2]), max(x[,2]))
```

and calculate $x_2$ as below to draw the boundary line:

$x_2 = - \dfrac{\theta_0 + \theta_1 x_1}{\theta_2} = - \dfrac{1}{\theta_2} (\theta_0 + \theta_1 x_1) $



````octave
% Initialize fitting parameters
n = size(xdata)(2);
theta = zeros(n+1, 1);
% Define the sigmoid function
g = inline('1.0 ./ (1.0 + exp(-z))'); 

% Newton's method
MAX_ITR = 7;
J = zeros(MAX_ITR, 1);

for i = 1:MAX_ITR
    % Calculate the hypothesis function
    z = X * theta;
    h = g(z);
   
    % Calculate gradient and hessian.
    % The formulas below are equivalent to the summation formulas
    % given in the lecture videos.
    grad = (1/m) .* X' * (h-Y);
    H = (1/m) .* X' * diag(h) * diag(1-h) * X;

    % Calculate J (for testing convergence)
    J(i) =(1/m)*sum(-Y.*log(h) - (1-Y).*log(1-h));

    theta = theta - H\grad;
end
% Display theta
theta
````

````octave
% Calculate the probability that a student with
% Score 20 on exam 1 and score 80 on exam 2
% will not be admitted
prob = 1 - g([1, 20, 80]*theta)
````

The probability that a student with a score of 20 on Exam 1 and 80 on Exam 2 will not be admitted to college is 0.668. 

````octave
% Plot J
figure
plot(0:MAX_ITR-1, J, 'o--', 'MarkerFaceColor', 'r', 'MarkerSize', 8)
xlabel('Iteration'); ylabel('J')
````
